```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('diabetes_data_upload.csv')
df
```

```python
df.shape
```

```python
df.replace({'Gender':{'Male':1, 'Female':0}}, inplace=True)
df.replace({'Polyuria':{'Yes':1, 'No':0}}, inplace=True)
df.replace({'sudden weight loss':{'Yes':1, 'No':0}}, inplace=True)
df.replace({'weakness':{'Yes':1, 'No':0}}, inplace=True)
df.replace({'Polyphagia':{'Yes':1, 'No':0}}, inplace=True)
df.replace({'Polydipsia':{'Yes':1, 'No':0}}, inplace=True)
df.replace({'Genital thrush':{'Yes':1, 'No':0}}, inplace=True)
df.replace({'visual blurring':{'Yes':1, 'No':0}}, inplace=True)
df.replace({'Itching':{'Yes':1, 'No':0}}, inplace=True)
df.replace({'Irritability':{'Yes':1, 'No':0}}, inplace=True)
df.replace({'delayed healing':{'Yes':1, 'No':0}}, inplace=True)
df.replace({'partial paresis':{'Yes':1, 'No':0}}, inplace=True)
df.replace({'muscle stiffness':{'Yes':1, 'No':0}}, inplace=True)
df.replace({'Alopecia':{'Yes':1, 'No':0}}, inplace=True)
df.replace({'Obesity':{'Yes':1, 'No':0}}, inplace=True)
df.replace({'class':{'Positive':1, 'Negative':0}}, inplace=True)
df
```

```python
df.describe()
```

```python
x=df.drop(columns=['class'],axis=1)
y=df['class']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.1,random_state=6969)
dtc_gini = DecisionTreeClassifier(criterion='gini')
```

```python
dtc_gini.fit(x_train,y_train)
```

```python
dtc_gini.score(x_test,y_test)
```

```python
dtc_entropy=DecisionTreeClassifier(criterion='entropy')
```

```python
dtc_entropy.fit(x_train,y_train)
```

```python
dtc_entropy.score(x_test,y_test)
```

```python
sc = StandardScaler()
sc.fit(x_train)
```

```python
x_train_sc=sc.transform(x_train)
x_test_sc=sc.transform(x_test)
```

```python
class_sc=DecisionTreeClassifier(criterion='gini')
class_sc.fit(x_train_sc, y_train)
class_sc.score(x_test_sc, y_test)
```

```python
class_en=DecisionTreeClassifier(criterion='entropy')
class_sc.fit(x_train_sc, y_train)
class_sc.score(x_test_sc, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```
